```python
# 13. Faça um programa que pergunte ao usuário se ele quer passar uma temperatura de Fahrenheit para Celsius ou de Celsius para Fahrenheit, e que, a partir da resposta do usuário, faça a devida conversão.


def fahrenheit_para_celsius(fahrenheit):
    celsius = (fahrenheit - 32) * 5/9
    return celsius

def celsius_para_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

def main():
    escolha = input("Você quer converter de Fahrenheit para Celsius (digite 'F') ou de Celsius para Fahrenheit (digite 'C')? ").upper()

    if escolha == 'F':
        fahrenheit = float(input("Digite a temperatura em Fahrenheit: "))
        celsius = fahrenheit_para_celsius(fahrenheit)
        print(f"A temperatura em Celsius é: {celsius:.2f}°C")
    elif escolha == 'C':
        celsius = float(input("Digite a temperatura em Celsius: "))
        fahrenheit = celsius_para_fahrenheit(celsius)
        print(f"A temperatura em Fahrenheit é: {fahrenheit:.2f}°F")
    else:
        print("Opção inválida. Por favor, escolha 'F' para Fahrenheit para Celsius ou 'C' para Celsius para Fahrenheit.")

if __name__ == "__main__":
    main()
```
