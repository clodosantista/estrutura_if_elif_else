```python
 # Exercício com funções - Calculadora em Python
 # Você foi designado para desenvolver uma calculadora em Python que permita aos usuários realizar diversas operações matemáticas, como soma, subtração, multiplicação, divisão e potenciação.
 # Para isso, você precisa implementar as seguintes funções:

#1. soma(a, b): Retorna a soma dos números a e b. 2.
#2. subtracao(a, b): Retorna a subtração do número b do número a .
#3. multiplicacao(a, b): Retorna o produto dos números a e b.
#4. divisao(a, b): Retorna o resultado da divisão do número a pelo número b. Certifique-se avisar quando ocorrer divisão por zero.
#5. potenciacao(a, b): Retorna o resultado de a elevado à potência de b.

# Agora, você deve implementar a função calculadora() que exibe um menu para o usuário escolher a operação desejada e solicitar os operandos apropriados. Em seguida, a função deve chamar a função correspondente para realizar a operação desejada e exibir o resultado.Aqui está um esqueleto do código para ajudá-lo a entender melhor o problema e começar a resolvê-o:
# Calculadora em Python #
```

```python
# funcao para exibir o menu calculadora def exibir_menu():
# print("escolha a operacao:")

def soma(a,b):
  return (a+b)

def subtracao(a,b):
  return(a-b)

def multiplicacao(a,b):
  return(a*b)

def divisao(a,b):
  return(a/b)

def potenciacao(a,b):
  return(a**b)

def exibir_menu():
  print ("Escolha a operação:")
  print ("1. Soma")
  print("2. Subtração")
  print("3. Multiplicação")
  print("4. Divisão")
  print("5. Potenciação")
```

```python
def calculadora():
  exibir_menu()
  escolha= input("Digite opção desejada: ")
  num1 = float(input("Digite o primeiro número: "))
  num2 = float(input("Digite o segundo número: "))

  if escolha == '1':
    print("Resultado: ", soma(num1,num2))

  elif escolha == '2':
    print("Resultado: ", subtracao(num1,num2))

  elif escolha == '3':
    print("Resultado: ", multiplicacao(num1,num2))

  elif escolha == '4':
    print("Resultado: ", divisao(num1,num2))

  elif escolha == '5':
    print("Resultado: ", potenciacao(num1,num2))

  else:
    print("Opção inválida, digite uma das opções disponíveis!")

calculadora()
```

```python

# Desenvolva um programa em pyhton iniciante que valide se um cpf é válido  ou não. O CPF é composto por 11 digitos e segue um padrão especifíco  de validação
def valida_cpf(cpf):
    if len(cpf) != 11:
        return False

    # Calcula e verifica primeiro dígito verificador
    soma = sum(int(cpf[i]) * (10 - i) for i in range(9))
    d1 = (soma * 10 % 11) % 10
    if d1 != int(cpf[9]):
        return False

    # Calcula e verifica segundo dígito verificador
    soma = sum(int(cpf[i]) * (11 - i) for i in range(10))
    d2 = (soma * 10 % 11) % 10
    if d2 != int(cpf[10]):
        return False

    return True

cpf = input("Digite o CPF: ")
if valida_cpf(cpf):
    print("CPF válido.")
else:
    print("CPF inválido.")
```

```python
def calcular_digitos_verificadores(cpf_base):
    cpf = [int(d) for d in cpf_base if d.isdigit()]
    if len(cpf) != 9:
        return "CPF inválido. Insira os primeiros 9 dígitos."

    def calcular_digito(cpf):
        soma = 0
        peso = 2
        for digito in reversed(cpf):
            soma += digito * peso
            peso += 1
        resto = soma % 11
        return 0 if resto < 2 else 11 - resto

    digito1 = calcular_digito(cpf)
    cpf.append(digito1)
    digito2 = calcular_digito(cpf)

    return digito1, digito2

cpf_base = "291.542.938"  # Insira os primeiros 9 dígitos do CPF
cpf_puro = "".join(filter(str.isdigit, cpf_base))

digito1, digito2 = calcular_digitos_verificadores(cpf_puro)
print(f"Os dígitos verificadores do CPF são: {digito1} e {digito2}")
```
